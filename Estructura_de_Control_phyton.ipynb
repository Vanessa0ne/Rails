#SINTAXIS BASICA.

```python
# Preguntar al usuario por un número
numero = int(input("Introduce un número: "))

# Verificamos si el número es par
resto = numero % 2  # Obtenemos el resto de la división del número entre 2

if resto == 0:
    print("El número es par.")
else:
    print("El número es impar.")
```

```python
import random

# Generar un número aleatorio entre 1 y 100
numeroPC = random.randint(1, 100)
adivine = False

while not adivine:
    # Preguntar al usuario por un número
    numeroUsuario = int(input("Adivina el número: "))

    if numeroPC == numeroUsuario:
        print("¡Felicitaciones! Has adivinado el número.")
        adivine = True
    else:
        if numeroUsuario > numeroPC:
            print("Número muy grande, prueba uno más pequeño.")
        else:
            print("Número muy pequeño, prueba uno más grande.")
```

#Deteccion de Manos.

Mediapipe es una biblioteca Open Source desarrollada por Google. Ésta facilita la implementación de soluciones avanzadas de visión artificial, con modelos pre-entrenados para detectar elementos. Viene con un conjunto de herramientas y algoritmos para trabajar en conjunto con OpenCV, lo que nos permite hacer muy fácil el reconocimiento y seguimiento de gestos, caras, manos, poses corporales y muchas cosas más.

Algunas cosas que podemos hacer con Mediapipe

Reconocimiento de gestos y manos: Interpretar gestos de las manos para controlar dispositivos o aplicaciones.
Detección facial: Identificar y seguir caras en videos en tiempo real.
Seguimiento de poses: Analizar movimientos corporales, o poses del cuerpo humano.
Realidad aumentada: Superponer objetos virtuales en el mundo real, en lo que llamamos aplicaciones de Realidad Aumentada (AR/Augmented Reality)
PD: Si no tienen cámara, o no quieren usarla, pueden reemplazar el índice de la cámara (en mi caso el 6) por una ruta a cualquier video (pueden probar con el “video.mp4” que está en el directorio de trabajo).

```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Hands and drawing utilities
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils

# Initialize video capture
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error al abrir la cámara")

with mp_hands.Hands(
        static_image_mode=False,
        max_num_hands=2,
        min_detection_confidence=0.5,
        min_tracking_confidence=0.5) as hands:
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Error al capturar el frame")
            break

        # Convert the BGR image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Process the image and detect hands
        results = hands.process(image)

        # Convert the image color back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Draw hand landmarks if detected
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image, hand_landmarks, mp_hands.HAND_CONNECTIONS)

        # Display the image
        cv2.imshow('Hand Detection', image)

        # Exit the loop when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the capture and destroy all windows
cap.release()
cv2.destroyAllWindows()
```

```python
!pip install mediapipe

import cv2
import numpy as np
import math
from pymodbus.client import ModbusTcpClient  # Updated import statement
import mediapipe as mp

# Initialize Modbus client
try:
    PLC = ModbusTcpClient('192.168.0.1')  # Replace with the correct IP address
    if PLC.connect():
        print("Connected to Modbus server.")
    else:
        print("Failed to connect to Modbus server.")
except Exception as e:
    print(f"Error connecting to Modbus server: {e}")

# Initialize MediaPipe Hands
mpHands = mp.solutions.hands
mpDrawing = mp.solutions.drawing_utils

# Example usage of MediaPipe and OpenCV
cap = cv2.VideoCapture(0)

with mpHands.Hands() as hands:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert the frame to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        # Process the image and detect hands
        results = hands.process(image)

        # Draw hand landmarks
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        if results.multi_hand_landmarks:
            for hand_landmarks in results.multi_hand_landmarks:
                mpDrawing.draw_landmarks(image, hand_landmarks, mpHands.HAND_CONNECTIONS)

        # Display the resulting frame
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

# Release the capture
cap.release()
cv2.destroyAllWindows()

# Close the Modbus client
if PLC:
    PLC.close()
```
