```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4]  # Representación numérica para el eje x
y = [1, 2, 3, 4]  # Representación numérica para el eje y
tamaños = [60, 50, 45, 70]  # Tamaños de las burbujas
etiquetas = ['Animales vacunados 1', 'Animales vacunados 2', 'Animales vacunados 3', 'Condominios']  # Etiquetas para el gráfico

# Crear gráfico de burbujas
plt.scatter(x, y, s=tamaños, color='purple', alpha=0.5)

# Añadir etiquetas
for i, etiqueta in enumerate(etiquetas):
    plt.annotate(etiqueta, (x[i], y[i]), fontsize=12, ha='center')

# Títulos y etiquetas de los ejes
plt.title('Gráfico de Burbujas')
plt.xticks(x, ['flor de poblacion', 'ex vertedero', 'la cima', 'condominios'])  # Asignar las etiquetas originales al eje x
plt.yticks(y, ['octuple', 'antirrabica', 'triplefelina', 'peluquera y estetica'])  # Asignar las etiquetas originales al eje y
plt.xlabel('Participación vacunación animales')
plt.ylabel('Operativo vacunatorio')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3]  # Representación numérica para los años (2022, 2023, 2024)
y = [90, 60, 60]  # Tasas de vacunación: 90% para canes, 60% para felinos
tamaños = [60, 50, 45]  # Tamaños de los puntos
etiquetas = ['2022', '2023', '2024']  # Etiquetas para los años
colores = ['purple', 'orange', 'green']  # Colores de los puntos

# Crear gráfico de dispersión
plt.figure(figsize=(10, 6))  # Aumentar el tamaño de la figura
scatter = plt.scatter(x, y, s=tamaños, color=colores, alpha=0.5)  # Ajustar alpha

# Añadir etiquetas
for i, etiqueta in enumerate(etiquetas):
    plt.annotate(etiqueta, (x[i], y[i]), fontsize=12, ha='center', va='bottom')  # Ajustar tamaño de fuente

# Títulos y etiquetas de los ejes
plt.title('Actividades Vacunatorias Rescatistas y Animalistas Independientes de Alto Hospicio, Chile')
plt.xticks(x, ['2022', '2023', '2024'])  # Asignar las etiquetas de años al eje x
plt.yticks([60, 90], ['60% Felinos', '90% Canes'])  # Asignar las etiquetas de vacunación al eje y
plt.xlabel('Años')
plt.ylabel('Tasa de Participación (%)')

# Añadir una cuadrícula para facilitar la lectura
plt.grid(True, linestyle='--', alpha=0.5)  # Cambiar el estilo y la opacidad de la cuadrícula

# Añadir una leyenda
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='purple', markersize=10, label='Vacunación Canes (90%)'),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=10, label='Vacunación Felinos (60%)')]
plt.legend(handles=handles, title='Vacunas', loc='upper left')

# Mostrar el gráfico
plt.tight_layout()  # Ajustar para mejorar la presentación
plt.show()
```
